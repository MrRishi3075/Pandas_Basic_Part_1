## Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.



#### DataFrame:

* A DataFrame is a two-dimensional, tabular data structure in Pandas that resembles a spreadsheet or SQL Table.
* It is essentially a collection of Series objects, where each Series represents a column of data.
* DataFrames can store data of different types (integers, strings, floats, etc.) within the same structure.
* You can think of a DataFrame as a container for multiple Series that share the same index.
* DataFrames are typically used for handling structured data, like datasets from CSV files or database tables.

#### Series:

* A Series is a one-dimensional array-like data structure in Pandas. It can be thought of as a single column of data.
* Unlike DataFrames, Series can only store data of one data type (e.g., all integers, all strings, etc.).
* Each element in a Series has an associated index, which can be used to access specific elements or align Series based on their indices.
* Series are often used for representing single columns or extracting a single column from a DataFrame.

```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
```

```python
# Creating a Series from a DataFrame column
age_series = df['Age']
print("\nSeries:")
print(age_series)
```

## Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

##### Pandas provides a wide range of functions for manipulating data in a DataFrame. Here are some common functions and methods you can use along with examples of when you might use them:

#### 1.head() and tail():
* Use head(n) to display the first n rows of a DataFrame and tail(n) to display the last n rows. This is helpful for a quick look at the data.

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df2.head(5)  # Display the first 5 rows
```

```python
df2.tail(3)  # Display the last 3 rows
```

#### 2.describe():

* Use describe() to get summary statistics of numeric columns in the DataFrame, such as count, mean, std, min, and max.


```python
df2.describe()
```

#### 3.info():

* Use info() to get a concise summary of the DataFrame, including data types and non-null counts for each column.

```python
df2.info()
```

#### 4.loc[] and iloc[]:

* loc[] allows you to access rows and columns by label, while iloc[] allows you to access rows and columns by integer index. These are useful for selecting specific subsets of data.

```python
df2
```

```python
df2.loc[4, 'Pclass']  # Access the 'Pclass' of the Fifth row (index 4)
```

```python
df2.iloc[:, 1:3]   # Access columns 1 and 2 for all rows
```

#### 5.groupby():

* Use groupby() to group data based on one or more columns and perform aggregate operations on them.

```python
df2.groupby('Sex')['Age'].mean()  # Calculate the mean age by gender
```

#### sort_values():
* Use sort_values() to sort the DataFrame by one or more columns in ascending or descending order.

```python
df2.sort_values(by='Age', ascending=False)  # Sort by age in descending order
```

#### 6.filter():
* Use filter() to select columns based on their names or a custom function.

```python
df2.filter(like='Nam')  # Select columns containing 'Nam' in their names
```

#### 7.drop():
* Use drop() to remove rows or columns from the DataFrame.

```python
df2.drop([0, 2])       # Remove rows with index 0 and 2
df2.drop(columns='Age') # Remove the 'Age' column
```

#### 8.apply():
* Use apply() to apply a function to each element, row, or column in the DataFrame.

```python
df2['Age'] = df2['Age'].apply(lambda x: x + 1)  # Increment each age by 1
```

```python
df2
```

#### 9.pivot_table():
* Use pivot_table() to create a pivot table from the DataFrame, which is helpful for summarizing data.

```python
pd.pivot_table(df2, values='Age', index='Sex', aggfunc='mean')  # Pivot table for mean age by gender
```

## Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

##### In the context of pandas, which is a popular data manipulation library in Python, here is the mutability status of the mentioned data structures:

#### Series: Mutable

* A pandas Series is mutable, which means you can change its elements after creation. You can assign new values to specific elements in a Series.

#### DataFrame: Mutable

* A pandas DataFrame is also mutable. You can modify its content, add or remove columns, and change values within the DataFrame.

#### Panel: Deprecated
* The Panel data structure has been deprecated in more recent versions of pandas (since version 0.25.0) and is no longer a recommended way to store and manipulate data. You should use DataFrames and MultiIndex DataFrames instead. If you still have old code that uses Panels, you should consider refactoring it to use DataFrames.

###### Keep in mind that while Series and DataFrames are mutable, they also offer methods for creating new objects with the desired changes rather than modifying the original objects in place. This immutability behavior is common in many data manipulation libraries, and it helps avoid unexpected side effects when working with data.







## Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Create two Series
series1 = pd.Series([1, 2, 3, 4, 5, 6], name='A')
series2 = pd.Series(['Apple', 'Banana', 'Cherry', 'Dates', 'Guvava', 'Grapes'], name='B')

# Combine the Series into a dictionary
data = {'ColumnA': series1, 'ColumnB': series2}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
